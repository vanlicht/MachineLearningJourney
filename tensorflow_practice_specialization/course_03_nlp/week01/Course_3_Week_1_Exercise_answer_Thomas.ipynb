```
labels = []
sentences = []
with open('/tmp/bbc-text.csv', mode='r', newline= '') as csvfile:
  csvfile = csv.reader(csvfile)
  next(csvfile)
  for row in csvfile:
    labels.append(row[0])
    #remove stop words
    #filter(), see the link https://www.programiz.com/python-programming/methods/built-in/filter
    clean_texts = filter(lambda x: x not in stopwords, row[1].split())
    sentences.append(' '.join(clean_texts))
```

```
tokenizer = Tokenizer(oov_token='<OOV>')
tokenizer.fit_on_texts(sentences)
word_index = tokenizer.word_index
sequences = tokenizer.texts_to_sequences(sentences)
padded = pad_sequences(sequences, padding='post')
```
